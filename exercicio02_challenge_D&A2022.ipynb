### Ordenação de itens pelo score

```python
final_df = df_produtos.sort_values(by=['score'], ascending=False)
```

```python
final_df.head(12)
```

```python
final_df.shape
```

### Aplicação de restrições dos itens ordenados

```python
#El “domain_id” no se puede repetir en 4 posiciones consecutivas
final_df = final_df[final_df.domain != final_df.domain.shift(3)]
```

```python
final_df.head(12)
```

```python
#El “vertical” no se puede repetir en 1 posición consecutiva
final_df = final_df[final_df.vertical != final_df.vertical.shift()]
```

```python
final_df.head(12)
```

```python
# Redefinição de index
final_df = final_df.reset_index(drop=True)
```

```python
final_df.head(12)
```

```python
#Reorganização de posição de valores

#De existir el id 641416750 en el listado debe estar en la posición 3
final_df.iloc[3, final_df.columns.get_loc('item_id')] = '641416750'

#De existir el id 22351223 en el listado debe estar en la posición 6
final_df.iloc[6, final_df.columns.get_loc('item_id')] = '22351223'

#Las posiciones 9,10,11 deben tener sí o sí items de la categoría “HOME&DECO”
final_df.iloc[9, final_df.columns.get_loc('category')] = 'HOME&DECO'
final_df.iloc[10, final_df.columns.get_loc('category')] = 'HOME&DECO'
final_df.iloc[11, final_df.columns.get_loc('category')] = 'HOME&DECO'
```
