**3. Similitud entre productos**

**Fonte: items_titles.csv**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy
import numpy as np
import seaborn as sns
import sklearn
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import Normalizer
from sklearn import metrics
from sklearn.cluster import KMeans, MiniBatchKMeans
import warnings
```

```python
df_titulos = pd.read_csv("data/items_titles_test.csv")
```

```python
df_titulos.sample(5)
```

```python
#Quantidade de produtos
len(df_titulos['ITE_ITEM_TITLE'].unique())
```

```python
#Conversão de coluna df em lista
df_titulos_col = df_titulos['ITE_ITEM_TITLE'].tolist()
```

```python
#Vetor com base na frequência (contagem) de cada palavra dos produtos
vectorizer = CountVectorizer(min_df = 1, stop_words = 'english')
dtm = vectorizer.fit_transform(df_titulos_col)
pd.DataFrame(dtm.toarray(),index=df_titulos_col,columns=vectorizer.get_feature_names()).head(10)
```

```python
#Combinação linear de palavras pelo componente LSA
#Análise das relações entre o conjunto de documentos e os termos (Análise Semântica Latente)
lsa = TruncatedSVD(2, algorithm = 'arpack')
dtm = dtm.asfptype()
dtm_lsa = lsa.fit_transform(dtm)
dtm_lsa = Normalizer(copy=False).fit_transform(dtm_lsa)
```

```python
#Redução dos dados em 2 dimensões
pd.DataFrame(lsa.components_,index = ["component_1","component_2"],columns =
vectorizer.get_feature_names())
```

```python
pd.DataFrame(dtm_lsa, index = df_titulos_col, columns = ["component_1","component_2"])
```

```python
#Criação da matriz de similaridade dos títulos de produtos 
similarity = np.asarray(numpy.asmatrix(dtm_lsa) * numpy.asmatrix(dtm_lsa).T)
pd.DataFrame(similarity,index=df_titulos_col, columns=df_titulos_col).head(10)
```

```python
#Atribuição da matriz em dataframe
df_titulos_matrix = pd.DataFrame(similarity,index=df_titulos_col, columns=df_titulos_col)
```

```python
#Tratamento do dataframe 
df_titulos_matrix['ITE_ITEM_TITLE'] = df_titulos_matrix.index
df_titulos_matrix_score = pd.melt(df_titulos_matrix.reset_index(), id_vars=['ITE_ITEM_TITLE'], var_name='ITE_ITEM_TITLE_S', value_name='Score_Similitud'). sort_values(['ITE_ITEM_TITLE'], ascending=False)
df_titulos_matrix_score = df_titulos_matrix_score.loc[df_titulos_matrix_score['ITE_ITEM_TITLE_S']!='index']
```
