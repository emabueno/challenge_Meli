```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
plt.style.use("ggplot")
```

```python
df_ofertas = pd.read_csv("data/ofertas_relampago.csv")
```

```python
df_ofertas.sample(10)
```

```python
print('O dataset possui',df_ofertas.shape[0], 'registros e', df_ofertas.shape[1], 'atributos.')
```

### Tratamento e normalização dos dados

```python
df_ofertas.columns
```

```python
df_ofertas.columns = ['data_inicio_oferta', 'datatime_inicio_oferta', 'datatime_fim_oferta',
                      'tipo_oferta', 'acoes_envolvidas', 'estoque_restante',
                      'valor_vendido', 'qtd_vendida', 'origem', 'tipo_pagt_envio',
                      'dom_domain_agg1', 'vertical', 'domain_id']
```

```python
df_ofertas.sample(5)
```

### Detecção de Outliers

```python
#Tratamento de Outliers

# Scatter plot
fig, ax = plt.subplots(figsize = (8,5))
ax.scatter(df_ofertas['valor_vendido'], df_ofertas['qtd_vendida'])
 
# x-axis label
ax.set_xlabel('(Valor Vendido)')
 
# y-axis label
ax.set_ylabel('Quantidade Vendida')
plt.show()
```

```python
#Tratamento dos valores discrepantes por pontuação de intervalo interquartil(IQR)
cols = ['acoes_envolvidas', 'estoque_restante','valor_vendido','qtd_vendida']

Q1 = df_ofertas[cols].quantile(0.25)
Q3 = df_ofertas[cols].quantile(0.75)
IQR = Q3 - Q1

df_ofertas_out = df_ofertas[~((df_ofertas[cols] < (Q1 - 1.5 * IQR)) |(df_ofertas[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
#Verificação de missings(Percentual)
df_ofertas_out.isna().sum()/df_ofertas_out.shape[0]*100
```

```python
df_ofertas_out['origem'].unique()
```

```python
#Eliminação dos dados de origem
df_ofertas_out = df_ofertas_out.drop(columns=['origem'],inplace=False)
```

**Os dados de origem são majoritáriamente nulos, correspondendo 70% da base, em que é composto apenas pela origem "A", dessa forma os missings foram retirados**

```python
# Substituição de missing de acordo com a média por grupo de produto
med_valor = df_ofertas_out.groupby('dom_domain_agg1')['valor_vendido'].transform('mean')
df_ofertas_out['valor_vendido'].fillna(med_valor, inplace=True)

med_qtd = df_ofertas_out.groupby('dom_domain_agg1')['qtd_vendida'].transform('mean')
df_ofertas_out['qtd_vendida'].fillna(med_qtd, inplace=True)
```

```python
#Tratamento de datas
df_ofertas_out['datatime_inicio_oferta'] = pd.to_datetime(df_ofertas_out['datatime_inicio_oferta'])
df_ofertas_out['datatime_fim_oferta'] = pd.to_datetime(df_ofertas_out['datatime_fim_oferta'])
```

```python
print('O dataset possui',df_ofertas_out.shape[0], 'registros e', df_ofertas_out.shape[1], 'atributos.')
```

**Com o tratamento de remoção de outliers, 6.892 registros foram retirados do dataframe**

### Distribuição de vertical dos produtos

```python
plt.figure(figsize=(8,6))
df_ofertas_out.vertical.value_counts().plot(kind='pie', autopct='%.2f%%')
plt.axis('equal');
```

**Com base em análises a vertical com maior concentração dos dados é App & Sports com 29.30%, seguindo Home & Industry com 22.32% e Eletrônicos com 16.91%**

### 10 grupos (dom_domain_agg1) de produtos mais vendidos

```python
plt.figure(figsize=(14,5))
df_ofertas_out.groupby("dom_domain_agg1")["qtd_vendida"].sum().sort_values(ascending=False).head(10).plot(kind='bar', color="#F5BC21")
```

**Dentre os 10 grupos de produtos mais vendidos os 3 primeiros foram Home & Decor, seguindo de Apparel Accesories e Computers**

### 3 grupos (dom_domain_agg1) de produtos menos vendidos

```python
plt.figure(figsize=(8,5))
df_ofertas_out.groupby("dom_domain_agg1")["qtd_vendida"].sum().sort_values(ascending=True).head(3).plot(kind='bar', color="#F5BC21")
```

**Os 3 grupos de produtos menos vendidos foram Musical Instruments, Antiques & Hobbies e Agro**

### 3 verticais com maior índice de vendas

```python
plt.figure(figsize=(8,5))
df_ofertas_out.groupby("vertical")["qtd_vendida"].sum().sort_values(ascending=False).head(3).plot(kind='bar', color="#72B6A1")
```

**As verticais de produtos com maior vendas foram App & Sports, Home & Industry e Eletrônicos**

### 5 grupos(domain) que mais tiveram ações envolvidas

```python
plt.figure(figsize=(8,5))
df_ofertas_out.groupby("dom_domain_agg1")["acoes_envolvidas"].sum().sort_values(ascending=False).head(3).plot(kind='bar', color="#C2B787")
```

**Dentre os 5 grupos de produtos com maior acões envolvidas Home & Decor foi o que teve maior quantidade**

### Concentração de vendas por vertical

```python
plt.figure(figsize=(16,5))
ax = sns.boxplot(x="vertical", y="valor_vendido", data=df_ofertas_out, palette="Set2", showfliers=False)
```

**De acordo com observações, as verticais de produtos que mais tiveram valores altos nas vendas foram Eletrônicos (CE) mais de 30 e  concentrações de valores em 15 à 20, T&B até 28 e concentrações em 12 e 18 e Home & Industry até 30 com concentrações mais baixas entre 8 à 16**<br>
**CPG, Others, App & Sports tiveram vendas de valores máximos de 15**


### Grupos(domain) que apresentaram valores maiores de vendas

```python
plt.figure(figsize=(8,5))
df_ofertas_out.groupby("dom_domain_agg1")["valor_vendido"].sum().sort_values(ascending=False).head(3).plot(kind='bar', color="#95A3C3")
```

**Grupos de produtos que tiveram valores de vendas mais altos foram Home & Decor, Eletronics e Apparel Accesories**

### Quantidade de vendas ao longo das ofertas

```python
plt.figure(figsize=(14,5))
df_ofertas_out.groupby(df_ofertas_out["data_inicio_oferta"])["qtd_vendida"].sum().plot(color='#347CB7')
plt.title("Quantidade de vendas ao longo das ofertas")
plt.xlabel("Data Início Ofertas")
plt.ylabel("Quantidade Produtos Vendidos")
plt.show()
```

**De acordo com análises, as vendas de produtos tiveram quedas em maio e queda maior no final de junho**<br>
**As vendas apresentaram picos na metadde de junho, início e final de julho**

### Correlação dos dados - Pearson

```python
sns.heatmap(df_ofertas_out.corr(),
            annot = True,
            fmt = '.2f',
            cmap='Blues')
plt.show()
```

**Os dados que mais tiveram correlação foram estoque restante e ações envolvidas com correlação forte de 0.91, quantidade vendida e valor vendido tiveram correlação fraca de 0.41**
